# 1. `Food Delivery` Dataset

This dataset contains food delivery times based on various influencing factors such as distance, weather, traffic conditions, and time of day. It offers a practical and engaging challenge for machine learning practitioners, especially those interested in logistics and operations research. Read and load the data from kaggle: [Food Delivery Dataset](https://www.kaggle.com/datasets/denkuznetz/food-delivery-time-prediction/data).

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("denkuznetz/food-delivery-time-prediction")

# Import data
import pandas as pd
data = pd.read_csv(path + "/Food_Delivery_Times.csv")
data.head()
```

**A.** What's the dimension of the data? Which variables are considered quantitative and which are qualitative?

> **Answer:**The dimension of a dataset refers to its number of rows and columns. Quantitative: variables that represent measurable quantities (numerical). Qualitative: variables that represent categories, groups, or labels (not meaningful to average).

**B. Qualitative variables:**

- Create statistical summary of qualitative columns.

- Create graphical representation of these qualitative columns to understand them better.

- Explain each column based on the stastical values and graphs.

```python
# B

print(f"The qualitative columns include: {list(data.select_dtypes(include = 'object').columns)}")

print(f"The quantitative columns include: {list(data.select_dtypes(exclude = 'object').columns)}")
```

```python
qual_vars = list(data.select_dtypes(include = 'object').columns)
for va in qual_vars:
  print(data[va].value_counts(normalize = True))
  print('\n')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

qual_data = ['Weather', 'Traffic_Level', 'Time_of_Day', 'Vehicle_Type']

for col in qual_data:
    plt.figure(figsize=(10, 4))

    # Bar plot
    plt.subplot(1, 2, 1)
    ax = sns.countplot(data, x=col)
    ax.bar_label(ax.containers[0])
    plt.title(f'{col} - Bar Plot')

    # Pie chart
    plt.subplot(1, 2, 2)
    data[col].value_counts().plot.pie(autopct='%1.1f%%')
    plt.title(f'{col} - Pie Chart')
    plt.ylabel("")

    plt.tight_layout()
    plt.show()
```

Qualitative Variables in the Food Delivery Dataset

1. Weather

Statistical Summary: There are 5 weather categories: Clear, Rainy, Foggy, Snowy, and Windy. The most common weather is Clear, accounting for 48.5% of records, followed by Rainy (21.0%), Foggy (10.6%), Snowy (10.0%), and Windy (9.9%).

Graphical Insight: Clear days dominate the dataset, while Windy days are the least frequent.

2. Traffic Level

Statistical Summary: Traffic is categorized into Low, Medium, and High. Medium traffic is most frequent at 40.2%, followed closely by Low traffic (39.5%), and High traffic (20.3%).

Graphical Insight: Low and Medium traffic levels are favorable for deliveries, whereas High traffic is less common and may slow deliveries.

3. Time of Day

Statistical Summary: Deliveries occur in Morning, Afternoon, Evening, and Night. The Morning has the highest proportion (31.8%), followed by Evening (30.2%), Afternoon (29.3%), and Night (8.8%).

Graphical Insight: Delivery activity is fairly consistent during Morning, Afternoon, and Evening, while Night deliveries are much lower.

4. Vehicle Type

Statistical Summary: Vehicles include Bike, Scooter, and Car. Bikes are the most common (50.3%), followed by Scooters (30.2%) and Cars (19.5%).

Graphical Insight: Bikes dominate delivery vehicles, with Cars being the least used.

**C. Quantitative variables:**

- Create statistical summary of quantiative columns.

- Create graphical representation of these quantitative columns to understand them better.

- Explain each column based on the stastical values and graphs.

- Are there any columns with outliers?

```python
quan_vars = list(data.select_dtypes(exclude = 'object').columns)
data[quan_vars].describe()
```

```python
for quan in quan_vars:
  plt.figure(figsize= (15,4))

  # histogram
  plt.subplot(1,3,1)
  hist = sns.histplot(data[quan])
  plt.title(f'{quan} - Histogram')
  plt.xlabel(quan)
  plt.ylabel("Frequency")

  #box plot
  plt.subplot(1,3,2)
  box_plot = sns.boxplot(x=data[quan])
  plt.title(f'{quan} - Box Plot')
  plt.xlabel(quan)
  # Add median label
  median = data[quan].median()
  plt.text(median, 0, f'Median: {median:.2f}', ha='center', va='bottom', color='red')

  #density

  plt.subplot(1,3,3)
  density = sns.kdeplot(data[quan], fill=True)
  plt.title(f'{quan} - Density Plot')
  plt.xlabel(quan)
  plt.ylabel("Density")

  plt.show()
```
